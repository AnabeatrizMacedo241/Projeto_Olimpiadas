```python
# bibliotecas
from selenium import webdriver
from time import sleep
from selenium.webdriver import Chrome
import pandas as pd
from time import sleep
import re
```

```python
op = webdriver.ChromeOptions()
url = 'https://olympics.com/pt/olympic-games/rio-2016/medals?utm_source=google&utm_medium=essence-paid&utm_campaign=google_cpc_br_bau_hybrid_portuguese&utm_content=google_cpc_br_bau_hybrid_pt_na_na_dynamic_na&gclid=Cj0KCQjw6s2IBhCnARIsAP8RfAhsvUp15BzR54l73j-2KFnIfoniNyBnaarfARq5SWf1SE2TdArJDmAaAngsEALw_wcB&gclsrc=aw.ds'
```

```python
def obter_driver():
    return webdriver.Chrome()
```

```python
def scraping(link):
    driver = obter_driver()
    driver.get(link)
    sleep(4)
    
    cookies= driver.find_element_by_xpath("""//*[@id="onetrust-accept-btn-handler"]""")
    cookies.click()  
    
    sleep(3) #clicar para ordenar
    
    paises_ = driver.find_elements_by_class_name('styles__CountryWrapper-sc-fehzzg-4.IzmaI')
    lst_paises = []
    for paises in paises_:
        lst_paises.append(paises.text)
    print(lst_paises)
    print(' ')
    
    medalhas = driver.find_elements_by_class_name('Medalstyles__Wrapper-sc-1tu6huk-0.jUTlVC')
    
    ouro = []
    prata = []
    bronze = []
    total = []
    txt_medalhas = []
    for texto_medalhas in medalhas:
        txt_medalhas.append(texto_medalhas.text)
    for i in range(0, len(txt_medalhas), 4):
        ouro.append(txt_medalhas[i])
        prata.append(txt_medalhas[i+1])
        bronze.append(txt_medalhas[i+2])
        total.append(txt_medalhas[i+3])
    display(ouro, prata, bronze, total)
    print(' ')
    
    # df.options_display_max_rows = 86
    pd.options.display.max_rows = 999
    col = []
    for i in range(1,len(lst_paises)+1):
        col.append(i)
    df = pd.DataFrame({'Colocação' : col, 'País' : lst_paises, 'Ouro' : ouro, 'Prata' : prata, 'Bronze' : bronze, 'Total' : total})
    display(df)
    print(' ')
    # df.to_csv('olimpiada_colocação2016_.csv', index= False, sep = ';', encoding="utf-8")
```

```python
scraping(url)
```

```python
idh = pd.read_csv('./Data/idh_global2014.csv', sep=';')

idh
```

```python
ranking_format = pd.read_csv('./Data/idh_global2014_format.xls', sep = ';')
# ranking_format = ranking_format.set_index('Colocação')
ranking_format['País'].iloc[7] = 'Austrália' 
ranking_format = ranking_format.drop([83,84,85])

ranking_format
```

```python

join_all = ranking_format.merge(idh, left_on='País', right_on='País')
join_all = join_all.sort_values(by='Ranking IDH Global')
join_all = join_all.rename_axis('index').reset_index()
join_all = join_all.drop(['index'], axis=1)
join_all['Total']= join_all['Total'].astype(int)
join_all['Colocação']= join_all['Colocação'].astype(int)
join_all['Ouro'] = join_all['Ouro'].astype(str).str.replace("-", "0").astype(int)
join_all['Prata'] = join_all['Prata'].astype(str).str.replace("-", "0").astype(int)
join_all['Bronze'] = join_all['Bronze'].astype(str).str.replace("-", "0").astype(int)
join_all['IDH'] = join_all['IDH'].astype(str).str.replace(",", ".").astype(float)

join_all
```

```python
ranking_pais = ranking_format[['País', 'Colocação']]

ranking_pais
```

```python
join_rank = ranking_pais.merge(join_all, left_on='País', right_on='País')
join_rank.drop('Colocação_x',axis=1, inplace=True)
join_rank = join_rank.rename(columns={'Colocação_y':'Colocação'})

display(join_rank)
# join_rank.info()
# join_rank.to_csv('./Data/idh_com_colocacao.csv', index=False)
```

```python
# Média de IDH dos primeiros 10 países
print(round(join_rank.loc[0:10, 'IDH'].mean(), 2))

# Média de IDH dos 10 países do meio
print(round(join_rank.loc[34:44, 'IDH'].mean(), 2))

# Média de IDH dos últimos 10 países
print(round(join_rank.loc[70:80, 'IDH'].mean(), 2))
```
