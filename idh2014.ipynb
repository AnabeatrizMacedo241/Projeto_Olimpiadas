```python
# bibliotecas
from selenium import webdriver
from time import sleep
from selenium.webdriver import Chrome
from selenium.webdriver.common.keys import Keys
import pandas as pd
```

```python
op = webdriver.ChromeOptions()
url = 'https://www.br.undp.org/content/brazil/pt/home/idh0/rankings/idh-global.html'
driver = Chrome(r"C:\Users\marin\Desktop\chrome_driver\chromedriver.exe", options=op)
driver.get(url)
```

```python
txt_table = []
head = driver.find_elements_by_tag_name('th')

for title in head:
    title = title.text
    txt_table.append(title)
    
lines = driver.find_elements_by_tag_name('td')

for content in lines:
    content = content.text
    txt_table.append(content)

txt_table
```

```python
txt_table = txt_table[0:567]
```

```python
ranking = []
pais = []
idh = []

for n in range(0, (len(txt_table)), 3):
    ranking.append(txt_table[n])
    idh.append(txt_table[n+2])
    pais.append(txt_table[n+1])
    
print(ranking, pais, idh)

```

```python
ranking = ranking[1:]
pais = pais[1:]
idh = idh[1:]
```

```python
df = pd.DataFrame({'Ranking IDH Global': ranking, 'País': pais, 'IDH':idh})
df = df.set_index('Ranking IDH Global')
df
```
